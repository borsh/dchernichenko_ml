```python
import tensorflow as tf
#tf.enable_eager_execution()
```

```python
from sklearn.datasets import fetch_mldata
mnist = fetch_mldata('MNIST original')
mnist
```

```python
X, y = mnist["data"], mnist["target"]
X.shape
```

```python
sess = tf.Session()
```

```python
#tf.enable_eager_execution()
dataset1 = tf.data.Dataset.from_tensor_slices(X)
dataset2 = tf.data.Dataset.from_tensor_slices(y)
iterator1 = dataset1.make_one_shot_iterator()
next_element1 = iterator1.get_next()
iterator2 = dataset2.make_one_shot_iterator()
next_element2 = iterator2.get_next()
```

```python
print(dataset1.output_shapes)
print(dataset2.output_shapes)
```

```python
value1 = sess.run(next_element1)
value2 = sess.run(next_element2)
print(value1, value2)
```

```python
X_train = mnist.train.images
X_test = mnist.test.images
y_train = mnist.train.labels.astype("int")
y_test = mnist.test.labels.astype("int")
```

```python
import tensorflow as tf

config = tf.contrib.learn.RunConfig(tf_random_seed=42) # not shown in the config

feature_cols = tf.contrib.learn.infer_real_valued_columns_from_input(X_train)
dnn_clf = tf.contrib.learn.DNNClassifier(hidden_units=[300,100], n_classes=10,
                                         feature_columns=feature_cols, config=config)
dnn_clf = tf.contrib.learn.SKCompat(dnn_clf) # if TensorFlow >= 1.1
dnn_clf.fit(X_train, y_train, batch_size=50, steps=40000)
```

```python
from sklearn.metrics import accuracy_score

y_pred = dnn_clf.predict(X_test)
accuracy_score(y_test, y_pred['classes'])
```
